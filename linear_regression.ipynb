```python
import matplotlib.pyplot as plt
```

```python
X = [30, 43, 25, 51, 40, 20]
Y = [2.5, 3.4, 1.8, 4.5, 3.2, 1.6]

m = len(X)

#init param a b
a = 0
b = 0
```

```python

#mse
def loss_func(a, b, X, Y):
    error = 0
    for (x, y) in zip(X, Y):
        error += (a * x + b - y) ** 2
    return error / (2 * m)

def deriv_a(a, b, X, Y):
    Ea = 0
    for (x, y) in zip(X, Y):
        Ea += (a * x + b - y) * x
    return Ea / m

def deriv_b(a, b, X, Y):
    Eb = 0
    for (x, y) in zip(X, Y):
        Eb += (a * x + b - y)
    return Eb / m

# batch Gradiant descent
def gradient_descent(a, b, X, Y, alpha=0.001, iterations=100):
    for i in range(iterations):
        Ea = deriv_a(a, b, X, Y)
        Eb = deriv_b(a, b, X, Y)
        
        # Update
        a -= alpha * Ea
        b -= alpha * Eb
        
        # print the progress every 10 iterations
        if i % 10 == 0:
            print(f"Iteration {i}: a = {a}, b = {b}, loss = {loss_func(a, b, X, Y)}")

    return a, b
```

```python

#run
a_final, b_final = gradient_descent(a, b, X, Y)

#plot
plt.scatter(X, Y, color="blue", label="Data Points")
x_vals = [min(X), max(X)]
y_vals = [a_final * x + b_final for x in x_vals]
plt.plot(x_vals, y_vals, color="red", label=f"y = {a_final:.2f}x + {b_final:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```
